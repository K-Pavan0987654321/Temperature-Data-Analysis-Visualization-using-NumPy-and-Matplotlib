```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Temperature Data Analysis Project
# -------------------------------

# Step 1: Raw temperature data (float type allows NaN)
temps = np.array([28, 30, 35, -99, 33, 29, 41, 27], dtype=float)

# Step 2: Inspect data
print("Original Data:", temps)
print("Shape:", temps.shape)
print("Data Type:", temps.dtype)

# Step 3: Data Cleaning (replace invalid values with NaN)
temps[temps < 0] = np.nan
print("\nAfter Cleaning (NaN for invalid):", temps)

# Step 4: Remove NaN values
clean_temps = temps[~np.isnan(temps)]
print("Clean Temperature Data:", clean_temps)

# Step 5: Statistical Analysis
print("\n--- Statistical Analysis ---")
print("Average Temperature:", np.mean(clean_temps))
print("Maximum Temperature:", np.max(clean_temps))
print("Minimum Temperature:", np.min(clean_temps))
print("Standard Deviation:", np.std(clean_temps))

# Step 6: Identify Hot Days (>35Â°C)
hot_days = clean_temps[clean_temps > 35]
print("\nHot Days (>35Â°C):", hot_days)

# Step 7: Sort temperatures
sorted_temps = np.sort(clean_temps)
print("\nSorted Temperatures:", sorted_temps)

# Step 8: Top 3 hottest temperatures
top_3 = sorted_temps[-3:]
print("Top 3 Hottest Temperatures:", top_3)

# Step 9: Convert Celsius to Fahrenheit
fahrenheit = (clean_temps * 9/5) + 32
print("\nTemperatures in Fahrenheit:", fahrenheit)

# -------------------------------
# Visualization Section
# -------------------------------

days = np.arange(1, len(clean_temps) + 1)

# Visualization 1: Line Plot (Temperature Trend)
plt.figure()
plt.plot(days, clean_temps)
plt.xlabel("Days")
plt.ylabel("Temperature (Â°C)")
plt.title("Daily Temperature Trend")
plt.show()

# Visualization 2: Bar Chart (Daily Comparison)
plt.figure()
plt.bar(days, clean_temps)
plt.xlabel("Days")
plt.ylabel("Temperature (Â°C)")
plt.title("Daily Temperature Comparison")
plt.show()

# Visualization 3: Histogram (Temperature Distribution)
plt.figure()
plt.hist(clean_temps)
plt.xlabel("Temperature Range (Â°C)")
plt.ylabel("Frequency")
plt.title("Temperature Distribution")
plt.show()

# -------------------------------
# Weekly Temperature Analysis (2D Data)
# -------------------------------

# Random weekly data: 7 days Ã 3 times (Morning, Afternoon, Night)
weekly_temps = np.random.randint(25, 45, size=(7, 3))
print("\nWeekly Temperature Data (Morning, Afternoon, Night):\n", weekly_temps)

# Column-wise average (time-based)
avg_time_temp = np.mean(weekly_temps, axis=0)
print("Average Morning, Afternoon, Night Temps:", avg_time_temp)

# Row-wise maximum (daily hottest temperature)
daily_max = np.max(weekly_temps, axis=1)
print("Daily Maximum Temperatures:", daily_max)

# Visualization 4: Multi-line Plot (Weekly Pattern)
days = np.arange(1, 8)

plt.figure()
plt.plot(days, weekly_temps[:, 0], label="Morning")
plt.plot(days, weekly_temps[:, 1], label="Afternoon")
plt.plot(days, weekly_temps[:, 2], label="Night")
plt.xlabel("Days")
plt.ylabel("Temperature (Â°C)")
plt.title("Weekly Temperature Pattern")
plt.legend()
plt.show()
```
